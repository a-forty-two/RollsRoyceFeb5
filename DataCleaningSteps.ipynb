```python
# Outlier detection -> box plot, z-score
# z-score = (data-mean) / std

data.describe().T
```

```python
stats = data.describe().T
mu = stats['mean']
sigma = stats['std']
```

```python
mu
```

```python
zscoredata = (data-mu) / sigma
zscoredata.head(10)
```

```python

o1 =  (zscoredata > 3) | (zscoredata < -3)
o1
```

```python
o1.info()
```

```python
for name in o1.columns:
  print(name + " : " + str(o1[name].isna().sum()))
```

```python
o2=o1.index[o1.notna().any(axis=1)]
o2
```

```python
o1['isOutlier'] = o1.sum(axis=1)
```

```python
o1.head()
```

```python
alloutliers = o1[ o1['isOutlier'] > 0]
alloutliers.index
```

```python
data2 = data.drop(index= alloutliers.index)
data2
```
